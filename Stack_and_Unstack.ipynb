```python
planets=sns.load_dataset('planets')
planets
```

```python
planets.info()
```

```python
x = planets.groupby(['method'])[['mass','distance','orbital_period']].agg(['mean','median','std','min','max'])
```

```python
x
```

### STACK() : The stack method is used to pivot a level of the column labels (DataFrame's columns) to the innermost level of a MultiIndex on the rows (DataFrame's index). It "stacks" the specified level(s) from the columns to the rows.

##### SYNTAX :  DataFrame.stack(level=-1, dropna=True)
##### LEVEL : It tells level or levels to be stacked 


```python
planets_stack = x.stack(level=0) # stack(level=0)- to stack the level with index 0 of the columns. 
planets_stack
```

```python
x.stack(level=1) # stack(level=1)- to stack the level with index 1 of the columns.
```

##### Unstack() :The unstack function is used to pivot or reshape a DataFrame by moving a level of the index to become a level of the columns. This operation is the reverse of the stack function.
##### unstack is particularly useful when you have a DataFrame with a MultiIndex on its rows, and you want to move one of these levels back to the columns.

##### SYNTAX : DataFrame.unstack(level=-1, fill_value=None)

```python
planets_stack.unstack(level=0)
```

```python
planets_stack.unstack(level=1)
```
